```python
import pandas as pd
import requests
```

```python
import chardet

with open('books.csv', 'rb') as f:
    result = chardet.detect(f.read(50000))  # analyze first 50KB
    print(result)
```

```python
books = pd.read_csv('books.csv', encoding='ISO-8859-1', on_bad_lines="skip", sep=";")
print(list(books.columns))
```

```python
books = books.drop(['Image-URL-S', 'Image-URL-M', 'ISBN', 'Publisher'], axis=1)
```

```python
books = books.rename(columns={"Year-Of-Publication": "year", "Book-Author" : "author", "Book-Title" : "title"
                              ,"Image-URL-L" : "image"})
books.head()
```

```python
books['description'] = ''
books.head()
```

```python
api = "https://www.googleapis.com/books/v1/volumes"

params = {
    "q" : f"intitle:L'Etranger",
    "maxResults": 1
}

response = requests.get(api, params=params)

response = response.json()

print(response["items"][0]["volumeInfo"]["description"])


descriptions = []
for title in books['title']:
    response = requests.get(api, params = {"q" : f"intitle:{title}", "maxResults" : 1})

    response = response.json()

    if "items" in response and "volumeInfo" in response["items"][0] and "description" in response["items"][0]["volumeInfo"]:
        
        description = response["items"][0]["volumeInfo"]["description"]
        print(description[0:10])

    else:
        description = ""
    
    descriptions.append(description)

books["description"] = descriptions
    

```

```python

books.loc[:len(descriptions) - 1, "description"] = descriptions
```
